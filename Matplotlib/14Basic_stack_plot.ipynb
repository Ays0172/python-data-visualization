```python
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]
Player1 = [1, 2, 3, 3, 4, 4, 4, 4, 5]
Player2 = [2, 3, 4, 4, 5, 5, 5, 5, 6]
Player3 = [3, 4, 5, 5, 6, 6, 6, 6, 7]

# Stack Plot
labels = ['Player1', 'Player2', 'Player3']
colors = ['#ff9999','#66b3ff','#99ff99']

plt.stackplot(minutes, Player1, Player2, Player3, labels=labels, colors=colors)

plt.title("My Awesome Stack Plot")
plt.tight_layout()
plt.legend(loc = "upper left")
plt.show()
```

Comparing work done by developers. They work a total of 8 hrs a day. Developer1 slowly handsover the work and fades away.

```python
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')

days = [1, 2, 3, 4, 5, 6, 7, 8, 9]
developer1 = [8,6,5,5,4,2,1,1,0]
developer2 = [0,1,2,2,2,4,4,4,4]
developer3 = [0,1,1,1,2,2,3,3,4]

# Stack Plot
labels = ['Developer1', 'Developer2', 'Developer3']
colors = ['#ff9999','#66b3ff','#99ff99']

plt.stackplot(days, developer1, developer2, developer3, labels=labels, colors=colors)

plt.title("Work undertaken by a developer")
plt.tight_layout()
plt.xlabel("Days")
plt.ylabel("Hours")
plt.legend(loc = (1, 0.7))
plt.show()
```

## Stack Plot with a constant Total

```python
import numpy as np
import matplotlib.pyplot as plt

#Sample Data
data = np.array([[10,20,30,40], [20,30,10,40], [30,10,20,40]])
# print("Data:\n", data)

# Calculate totals for each category
totals = np.sum(data, axis = 0)
# print("Totals:\n", totals)

# Calculate the percentages for each value in each category
percentages = data / totals[np.newaxis, :] * 100
# print("Percentages:\n", percentages)

# Create a stacked plot
fig, ax = plt.subplots()        # Works same as plt

ax.stackplot(range(4), percentages, labels = ['Category 1', 'Category 2', 'Category 3'])
# @e use range(4) as it will take only first three values!
ax.legend(loc = (1,0.7))
ax.set_title('Stacked Plot with Percentages')
ax.set_ylabel('Percentage (%)')
ax.set_xlabel('Categories')
plt.show()
```
