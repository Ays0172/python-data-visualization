3D Plots

```python
import pandas as pd
import numpy as np
import cufflinks as cf
import plotly.express as px

# Create a random dataset

 # For reproducibility
df = pd.DataFrame({
    'x': np.random.rand(100),
    'y': np.random.rand(100),
    'z': np.random.rand(100)
})

# Create a 3D scatter plot
fig = px.scatter_3d(df, x='x', y='y', z='z')
fig.show()
```

Another Example using IRIS Dataset

```python
import cufflinks as cf
import plotly.express as px

# Load the iris dataset
df = px.data.iris()

# Create a 3D scatter plot
fig = px.scatter_3d(df, x='sepal_width', y='sepal_length', z='petal_length', color='species')
fig.show()
```

## Spread Plot & Hist Plot

```python
import plotly.express as px

# Load the tips dataset
df = px.data.tips()

# Create a Spread Plot
fig = px.scatter(df, x='total_bill', y='tip',  color='sex')
fig.show()

# Create a Histogram Plot
fig = px.histogram(df, x='total_bill', nbins = 30,  color='sex')
fig.show()
```

## Bubble Plot and Heatmap

```python
import pandas as pd
import numpy as np
import plotly.express as px
import cufflinks as cf
cf.go_offline()     # For offline use

# Create a sample data frame
np.random.seed(110) # For reproducibility
df = pd.DataFrame({
    'x': np.random.rand(100),
    'y': np.random.rand(100),
    'color': np.random.rand(100),
    'size': np.random.randint(1, 4, size = 100) # Random sizes between 1 and 3, a total of 100 points
})

# Bubble Plot
fig = px.scatter(df, x='x', y='y', size='size', color='color', hover_name=df.index)
fig.show()

# Heatmap
heatmap_data = df.pivot_table(values='size', index='y', columns='x')
fig = px.imshow(heatmap_data, color_continuous_scale = 'Blues')
fig.show()
```
