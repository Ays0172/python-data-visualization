### Stem Plot  
used to display both the shape and the distribution

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,]
y = [2,2,6,7,3,8]

plt.stem(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,]
y = [2,2,6,7,3,8]

plt.stem(x,y, linefmt = '-')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,]
y = [2,2,6,7,3,8]

plt.stem(x,y, linefmt = '-.', markerfmt='r^')        #  TO change shape and color; 'r' for color '^' for shape
plt.show()
```

To change the baseline; "bottom"

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,]
y = [2,2,6,7,3,8]

plt.stem(x,y, linefmt = '-.', markerfmt='ro', bottom = 3)
plt.show()
```

To change the color of the baseline

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,]
y = [2,2,6,7,3,8]

plt.stem(x,y, linefmt = '-.', markerfmt='ro', bottom = 3, basefmt = 'g')
plt.show()
```

Addying a label

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,]
y = [2,2,6,7,3,8]

plt.stem(x,y, linefmt = '-.', markerfmt='ro', bottom = 2, basefmt = 'r', label = 'python')
plt.legend()
plt.show()
```

### For better labelling

stem_container = plt.stem(x, y)  

The stem_container contains:  

stem_container[0] → the stem lines (the vertical lines from the baseline up to each data point).  

stem_container[1] → the markers (the symbols at the top of each stem line, like r^ in your case).  

stem_container[2] → the baseline (a horizontal line at the value of bottom).

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 6, 7, 3, 8]

stem_container = plt.stem(x, y, linefmt='-.', markerfmt='r^', bottom=1, basefmt='c')
stem_container[0].set_label('python')  # stemlines
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 6, 7, 3, 8]

stem_container = plt.stem(x, y, linefmt='-.', markerfmt='r>', bottom=1, basefmt='c', orientation = 'horizontal')
stem_container[0].set_label('python')  # stemlines
plt.legend()
plt.show()
```

Setting the location for the legend

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 6, 7, 3, 8]

stem_container = plt.stem(x, y, linefmt='-.', markerfmt='r>', bottom=1, basefmt='c', orientation = 'horizontal')
stem_container[0].set_label('python')  # stemlines
plt.legend(loc = (0.7, 0.1))
plt.show()
```
