### Connecting Data Points by Line

```python
import matplotlib.pyplot as plt

#  Example Data Points
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a line plot
plt.plot(x, y, 'o-.', label = "Data Points")

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Connecting Data Points by Line")

# Add legend
plt.legend()

# Show the plot
plt.show()
```

#### Changing the Formats of the dates


```python
import datetime
# Example date string in the original format
date_string = "2023-10-01"

# Convert the string to a datetime object
original_date = datetime.datetime.strptime(date_string, "%Y-%m-%d")

# Convert the datetime object to a different string format
new_date_string = original_date.strftime("%d/%m/%Y")

# Print the original date
print("Original date: ", original_date)
print("\n New date string: ", new_date_string)
```

# OR

```python
from datetime import datetime
# Example date string in the original format
date_string = "2023-10-01"

# Convert the string to a datetime object
original_date = datetime.strptime(date_string, "%Y-%m-%d")      # Removes the need to write "datetime.datetime.strptime"

# Convert the datetime object to a different string format
new_date_string = original_date.strftime("%d/%m/%Y")
new_date_string1 = original_date.strftime("%B %d, %Y")
new_date_string2 = original_date.strftime("%Y-%m-%d")
new_date_string = original_date.strftime("%b %d, %Y")


# Print the original date
print("Original date: ", original_date)
print("New date string: ", new_date_string)
print("New date string 1: ", new_date_string1)
print("New date string 2: ", new_date_string2)
print("New date string 3: ", new_date_string)
```

### Covert string dates using the .to_datetime() pandas method

```python
import pandas as pd

# Create a pandas Dataframe with a column of date strings
df = pd.DataFrame(
    {
        "Date": [
            "2023-10-01",
            "2023-10-02",
            "2023-10-03",
            "2023-10-04",
            "2023-10-05"
        ]
    }
)

# Convert the 'date' column to datetime objects using to_datetime()
df['Date'] = pd.to_datetime(df['Date'])

# Print the resulting DataFrame
print(df)
```

#### Unstructured Date formats  
We can easily read fromboth excel and csv files as well

```python
import pandas as pd

# Create a pandas Dataframe with a column of date strings
df = pd.DataFrame(
    {
        "Date": [
            "20231001",
            "20231002",
            "20231003",
            "20231004",
            "20231005"
        ]
    }
)

# Convert the 'Date' column to datetime objects using to_datetime() with format parameters
df['Date'] = pd.to_datetime(df['Date'], format='%Y%m%d')

# Print the resulting DataFrame
print(df)
```

```python
import pandas as pd
import numpy as np

# Create a pandas Dataframe with a column of date strings, icluding a missing value
df = pd.DataFrame(
    {
        "Date": [
            "20231001",
            "20231002",
            "None",
            "20231004",
            "20231005"
        ]
    }
)

# df
# Convert the 'date'column to datetime objects using to_datetime() with errors parameter
df['Date'] = pd.to_datetime(df['Date'], format='%Y%m%d', errors='coerce')
# df

# Fill in missing values with a default date
df['Date'] = df['Date'].fillna(pd.to_datetime('2023-01-01'))

# print the resulting DataFrame
print(df)
```
