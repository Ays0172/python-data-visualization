##### Addying a Median and Standard Deviation

```python
# Defining and manipulating bins by providing the desired number of bins

import numpy as np
import matplotlib.pyplot as plt

# Data in a Numpy Array
data = np.array([189, 185, 195, 149, 189, 147, 154,
          165, 172, 178, 182, 190, 175, 160, 167, 153,
          153, 158, 162, 154, 144, 176, 157, 176, 159,
          182, 181, 166, 167, 156, 145, 150, 155, 163,
          163, 170, 175, 180, 185, 190, 195, 200, 205, 133, 
          140, 145, 150, 155, 160, 165, 170, 175, 180, 185, 190, 195, 200, 205])

# Plotting the Histogram
plt.hist(data, bins=50, color='c', edgecolor='black', align='left', linewidth=1)
mean = np.mean(data)
median = np.median(data)
mode = np.bincount(data).argmax()       # Calculates Mode for the selected data
sd = np.std(data)
plt.axvline(mean, color='r', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(median, color='g', linestyle='dashed', linewidth=1, label='Median')
plt.axvline(mode, color='b', linestyle='dashed', linewidth=1, label='Mode')
plt.axvline(mean + sd, color='y', linestyle='dashed', linewidth=1, label='+ Standard Deviation')
plt.axvline(mean - sd, color='y', linestyle='dashed', linewidth=1, label='- Standard Deviation')
plt.legend()
plt.show()
```
