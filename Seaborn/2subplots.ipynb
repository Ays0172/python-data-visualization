```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the example tips dataset
tips = sns.load_dataset("tips")
print(tips)

# Create a figure with 2 subplots side-by-side
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Plot the first subplot - distribution plot of total bill
sns.histplot(ax=axes[0], data=tips, x="total_bill", kde=True, color="red", alpha=0.5)

# Set the title and x/y axis labels for the first subplot
axes[0].set_title("Distribution of Total Bill")
axes[0].set_xlabel("Total Bill")
axes[0].set_ylabel("Count")

# Plot the second subplot - scatter plot of total bill vs tip with hue based on time of day
sns.scatterplot(ax=axes[1], data=tips, x="total_bill", y="tip", hue="time")

# Set the title and x/y axis labels for the second subplot
axes[1].set_title("Total Bill vs Tip")
axes[1].set_xlabel("Total Bill")
axes[1].set_ylabel("Tip")

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.3)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the example tips dataset
tips = sns.load_dataset("tips")

# Create a figure with 4 subplots (2 rows, 2 columns)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Plot the first subplot - histogram with KDE of total bill
sns.histplot(ax=axes[0][0], data=tips, x="total_bill", kde=True, color="red", alpha=0.5)
axes[0][0].set_title("Distribution of Total Bill")
axes[0][0].set_xlabel("Total Bill")
axes[0][0].set_ylabel("Count")

# Plot the second subplot - violin plot of smoker vs tip
sns.violinplot(ax=axes[0][1], data=tips, x="smoker", y="tip", hue="sex", split=True)
axes[0][1].set_title("Tip by Smoker Status")
axes[0][1].set_xlabel("Smoker")
axes[0][1].set_ylabel("Tip")

# Plot the third subplot - strip plot of total bill vs tip
sns.stripplot(ax=axes[1][0], data=tips, x="smoker", y="tip", hue="time")
axes[1][0].set_title("Tip by Smoker Status")
axes[1][0].set_xlabel("Smoker")
axes[1][0].set_ylabel("Tip")

# Plot the fourth subplot - bar plot of total bill vs tip
sns.barplot(ax=axes[1][1], data=tips, x="sex", y="tip", hue="time")
axes[1][1].set_title("Tip by Gender")
axes[1][1].set_xlabel("Gender")
axes[1][1].set_ylabel("Tip")

# Adjust layout
plt.tight_layout()
plt.show()
```
