Joint Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the example tips dataset
tips = sns.load_dataset("tips")

# Create a joint plot of total bill and tip
sns.jointplot(data=tips, x="total_bill", y="tip", kind="scatter")

# Show the plot
plt.show()
```

Pair Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the example tips dataset
tips = sns.load_dataset("tips")

# Create a pair plot with hue based on sex
sns.pairplot(data=tips, hue='sex')

# Show the plot
plt.show()
```

Regression Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the example tips dataset
tips = sns.load_dataset("tips")

# Create a pair plot with hue based on sex
sns.regplot(data=tips, x = 'total_bill', y = 'tip')  # hue='sex' - NOT SUPPORTED!

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the example tips dataset
tips = sns.load_dataset("tips")

# Use lmplot for regression with hue support
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')

# Show the plot
plt.show()
```
