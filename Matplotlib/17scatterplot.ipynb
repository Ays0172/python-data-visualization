### Changing the size of the Dots

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating Random Data
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

# Creating a scatter plot with custom dot size
plt.scatter(x, y, s=50, alpha=0.5)

# Add Titles and Labels
plt.title("Scatter Plot with Custom Dot Size")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

# Show the Plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating Random Data
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

# Assign random sizes to each point
sizes = np.random.randint(10, 100, size = 100)   # Random sizes for each point

# Creating a scatter plot with custom dot size
plt.scatter(x, y, s=sizes, alpha=0.5)

# Add Titles and Labels
plt.title("Scatter Plot with Custom Dot Size")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

# Show the Plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating Random Data
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

# Define a condition for dot size
condition = x > 0  # Example condition: x values greater than 0

# Create an array of dot sizes based on the condition
sizes = np.where(condition, 120, 20)  # Larger dots for positive x values
## Alternatively could have written as
# sizes = np.where(x > 0, 100, 50)

# Creating a scatter plot with custom dot size based on condition
plt.scatter(x, y, s=sizes, alpha=0.5)

# Add Titles and Labels
plt.title("Scatter Plot with Custom Dot Size")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

# Show the Plot
plt.show()
```

### Changing colors of markers

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate Random Data
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

# Define a condition for dot size
condition = x > 0  # Example condition: x values greater than 0

# Create an array of dot sizes based on the condition
colors = np.where(condition, 'blue', 'red')  # Blue color for positive x values
#print(colors)

# Creating a scatter plot with custom dot size based on condition
plt.scatter(x, y, c=colors, alpha=0.5)

# Add Titles and Labels
plt.title("Scatter Plot with Custom Dot Size")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

# Show the Plot
plt.show()
```

How to use Color Map (cmap)

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate Random Data
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
z = np.random.normal(1, 0.5, 100)   # Third variable to determine marker color

#print(z)

# Create scatter plot with marker color determined by z values
plt.scatter(x, y, c = z, cmap='cool')    # virdis, magama, cool, hot

# Add colorbar
plt.colorbar()

# Add Titles and Labels
plt.title("Scatter Plot with Custom Marker colors based on Z values")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

# Show the Plot
plt.show()
```

Addying edges to the dots

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate Random Data
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

# Create Scatter Plots with eges
plt.scatter(x, y, color = 'green', edgecolor='black', linewidth=1)

# Add Titles and Labels
plt.title("Scatter Plot with Edges")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

# Show the Plot
plt.show()
```

Dynamic edges

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate Random Data
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
z = np.random.normal(1, 0.5, 100)   # Third variable to determine marker color

# Create a scatter plot with edges
plt.scatter(x, y, c = z, cmap = 'viridis', edgecolors = ['red', 'green', 'yellow'], linewidths = [1, 3, 5])

# Add a colorbar
plt.colorbar()

# Add Titles and Labels
plt.title("Scatter Plot with Custom Marker Edges")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

# Show the Plot
plt.show()
```
