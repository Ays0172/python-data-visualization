### Line Plot

```python
import cufflinks as cf
import plotly.express as px
import pandas as pd

# Load a sample Data
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv')

fig = px.line(df, x = 'Date', y = 'AAPL.High', title = 'AAPL High Prices')
fig.show()
```

### Scatter Plot

```python
import cufflinks as cf
import pandas as pd
import plotly.express as px

# Load sample data
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv')

# Create line plot with Plotly Express
fig = px.scatter(df, x='Date', y='AAPL.High', title='AAPL High Prices')
fig.show()
```

### Bar Graph

```python
import cufflinks as cf
import pandas as pd
import plotly.express as px

# Load sample data
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv')

# Create bar plot with Plotly Express
fig = px.bar(df, x='Date', y='AAPL.High', title='AAPL High Prices')
fig.show()
```

### Box Plot & Area Plot

```python
import cufflinks as cf
import pandas as pd
import plotly.express as px

# Load sample data
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv')

# Create box plot with Plotly Express
fig = px.box(df, x='Date', y='AAPL.High', title='AAPL High Prices', template='plotly_dark')
fig.show()
```

### Area Plot

```python
import cufflinks as cf
import pandas as pd
import plotly.express as px

# Load sample data
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv')

# Create area plot with Plotly Express
fig = px.area(df, x='Date', y='AAPL.High', title='AAPL High Prices', template='plotly_dark')
fig.show()
```

```python
import plotly.express as px
import pandas as pd

# Create Sample Data
data = pd.DataFrame({
    'x' : [1, 2, 3, 4, 5],
    'y1' : [2, 4, 1, 5, 3],
    'y2' : [1, 3, 2, 4, 2] 
})

# Create Area Plot
fig = px.area(data, x='x', y=['y1', 'y2'], title='Area Plot Example', template = 'plotly_dark')

# Show Plot
fig.show()
```
