```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Plot')
plt.show()
```

## Changing figuresize

```python
f = plt.figure(figsize=(2, 4))
plt.plot(x,y)
plt.show()
```
