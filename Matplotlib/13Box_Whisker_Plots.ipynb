### Boxplot

```python
import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50, 60, 70]
plt.boxplot(x)
plt.show()
```

#### Styling of Boxplots

```python
import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50, 60, 70]
plt.boxplot(x, notch = True, vert = False, widths = 0.5, labels = ['UK'], showmeans = True) #, patch_artist = True; Green spot shows mean
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50, 60, 70, 130]
plt.boxplot(x, notch = True, vert = False, widths = 0.5, labels = ['UK'], showmeans = True, whis = 3.5) # To include outlier within 3.5 * IQR
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50, 60, 70, 130]
plt.boxplot(x, notch = True, vert = False, widths = 0.5, labels = ['UK'], showmeans = True, sym = 'g+') # Green plus signs for outliers
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50, 60, 70, 130]
plt.boxplot(x, notch = False, vert = True, widths = 0.5, labels = ['UK'], showmeans = True, sym = 'g+', 
            boxprops = {'color': 'red'},  # Red box for the boxplot
            capprops = {'color': 'blue'}, # Blue caps for the whiskers
            whiskerprops= {'color': 'yellow'} # Yellow whiskers for the boxplot
            )
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50, 60, 70, 130]
x1 = [15, 30, 45, 60, 75, 90, 105, 120]


plt.boxplot([x, x1], notch = False, vert = True, widths = 0.5, labels = ['UK', 'US'], showmeans = True, sym = 'g+', 
            boxprops = {'color': 'red'},  # Red box for the boxplot
            capprops = {'color': 'blue'}, # Blue caps for the whiskers
            whiskerprops= {'color': 'yellow'} # Yellow whiskers for the boxplot
            )
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [10, 20, 30, 40, 50, 60, 70, 130]
x1 = [15, 30, 45, 60, 75, 90, 105, 120]
x2 = [20, 40, 60, 80, 100, 120, 140, 160]

y = [x, x1, x2]

plt.boxplot(y, notch = False, vert = True, widths = 0.5, labels = ['UK', 'US', 'CA'], showmeans = True, sym = 'g+', 
            boxprops = {'color': 'red'},  # Red box for the boxplot
            capprops = {'color': 'blue'}, # Blue caps for the whiskers
            whiskerprops= {'color': 'yellow'} # Yellow whiskers for the boxplot
            )
plt.show()
```
