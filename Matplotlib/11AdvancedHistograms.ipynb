```python
# To change the specific color of a bin for highlighting it

n, bins, patches = plt.hist([blood_sample_men, blood_sample_women], bins=[80, 100, 125, 150], color=['blue', 'pink'], alpha=0.7, label=['Men', 'Women'])
plt.legend()
plt.xlabel('Blood Sugar Levels')
plt.ylabel('Number of patients')
plt.title("Blood Level Chart")

# Highlight the patches with maximum values in cyan
plt.setp(patches[1][2], color='cyan')  # Change the color of the third bin of the second histogram
plt.setp(patches[0][0], color='cyan')  # Change the color of the first bin of the first histogram

# Highlight the patches with maximum values in red
plt.setp(patches[0][2], color='red')  # Change the color of the third bin of the first histogram
plt.setp(patches[1][1], color='red')  # Change the color of the second bin of the second histogram

plt.show()
```
