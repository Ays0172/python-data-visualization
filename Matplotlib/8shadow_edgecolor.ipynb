##### Changing the line color btw the edges

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = [ "Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, wedgeprops = { 'linewidth' : 2 , 'edgecolor' : 'white' } )
plt.show()
```

##### For changing the color of the slicers  

Create a set of colors  
colors = ['#4f6272', '#B7C3f3', '#dd7596', '#8eb897' ]

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = [ "Apples", "Bananas", "Cherries", "Dates"]

#Create a set of colors  
colors = ['#4f6272', '#B7C3f3', '#dd7596', '#8eb897' ]

plt.pie(y, labels = mylabels, wedgeprops = { 'linewidth' : 2 , 'edgecolor' : 'white' }, colors = colors )
plt.show()
```

To add shadows

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = [ "Apples", "Bananas", "Cherries", "Dates"]

#Create a set of colors  
colors = ['#4f6272', '#B7C3f3', '#dd7596', '#8eb897' ]

plt.pie(y, labels = mylabels, wedgeprops = { 'linewidth' : 2 , 'edgecolor' : 'white' }, colors = colors, shadow = 'true' )
plt.show()
```
