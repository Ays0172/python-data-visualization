##### Addying Legends to the pie chart

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = [ "Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```

Addying Title in Piechart and Legend

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = [ "Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.title('Fruits Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

# Set figure size before plotting
plt.figure(figsize=(8, 8))

# Create pie chart
plt.pie(y, labels=mylabels)

# Add title and legend
plt.title('Fruits Distribution')
plt.legend(title="Fruit names", loc="upper left")

# Show the plot
plt.show()
```

##### If we want to shwow the %ages  

autopct = "%1.1f%%'

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

# Set figure size before plotting
plt.figure(figsize=(8, 8))

# Create pie chart
plt.pie(y, labels=mylabels, autopct='%1.1f%%')      # `.1f means 1 digit after decimal, .2f means 2 digits and so on...

# Add title and legend
plt.title('Fruits Distribution')
plt.legend(title="Fruit names", loc="upper left")

# Show the plot
plt.show()
```

##### Rotating the pie chart  
startangle = 90, 180, 270, ...

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = [ "Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 120)

plt.show()
```
