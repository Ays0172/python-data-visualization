```python
import datetime

# Get the current date and time
now = datetime.datetime.now()

# Print the current date and time
print("Current date and time: ", now)
```

```python
import datetime

# Get the current date and time
now = datetime.datetime.now()

# Format the date and time output
formatted_date_time = now.strftime("%Y-%m-%d %H:%M:%S")
formatted_date_time01 = now.strftime("%Y/%m/%d %H:%M:%S")   # Styling can be changed, as desired

# Print the formatted date and time
print("Formatted date and time: ", formatted_date_time)
print("Formatted date and time (alternative): ", formatted_date_time01)
```

##### Calculating the time assed between two dates

```python
import datetime

# Create two dates
date1 = datetime.datetime(2023, 10, 1)
date2 = datetime.datetime(2023, 10, 15)

# Calculate the difference between the two dates
date_difference = date2 - date1

# Print the difference
print("Difference between dates: ", date_difference)
```
